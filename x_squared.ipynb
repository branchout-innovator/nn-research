```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generate training data
X = np.linspace(-10, 10, 200).reshape(-1, 1)
y = X**2

# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(X, y, epochs=1000, verbose=0)

# Generate predictions
X_test = np.linspace(-15, 15, 300).reshape(-1, 1)
predictions = model.predict(X_test)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Training data')
plt.plot(X_test, predictions, 'r', label='Predictions')
plt.plot(X_test, X_test**2, 'g--', label='True function')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Neural Network Approximation of x^2')
plt.show()

# Print the final loss
print(f"Final loss: {history.history['loss'][-1]:.6f}")
```

```python
import csv
import numpy as np

# Generate 1000 evenly distributed random numbers between -10 and 10
numbers = np.linspace(-10, 10, 1000)
np.random.shuffle(numbers)

# Calculate the squared values
squared_values = numbers ** 2

# Create the CSV file
with open('random_numbers_and_squares.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Number', 'Squared Value'])  # Write header
    for num, square in zip(numbers, squared_values):
        writer.writerow([num, square])

print("CSV file 'random_numbers_and_squares.csv' has been created.")
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generate training data
X = np.linspace(0, 100, 200).reshape(-1, 1)
y = np.sqrt(X)

# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(X, y, epochs=1000, verbose=0)

# Generate predictions
X_test = np.linspace(0, 130, 300).reshape(-1, 1)
predictions = model.predict(X_test)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Training data')
plt.plot(X_test, predictions, 'r', label='Predictions')
plt.plot(X_test, np.sqrt(X_test), 'g--', label='True function')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Neural Network Approximation of x^2')
plt.show()

# Print the final loss
print(f"Final loss: {history.history['loss'][-1]:.6f}")
```
